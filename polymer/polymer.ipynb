```python
import numpy as np
import os
from rdkit.Chem import AddHs, MolFromSmiles, MolToXYZBlock, AllChem

def OrcaWrapper(SMILES, theory, basis, *args, parallel=None):
    header = [
        "! RIJCOSX RI-B2PLYP D3BJ def2-tzvpp def2-tzvpp/C TIGHTSCF PAL8", 
        "! TightOpt NumFreq", 
        "* xyz 0 1" 
        ]

    if not os.path.exists(SMILES+".xyz"):
        mol = AddHs(MolFromSmiles(SMILES))
        AllChem.EmbedMolecule(mol)
        atom = MolToXYZBlock(mol).split("\n\n")[1]
    else:
        with open(SMILES+".xyz","r") as f:
            # file read skip first 2 lines
            atom = f.read().split("\n",2)[2]
    header.append(atom)
    header.append('*')

    return header

# SMILESS = [ "O","PET", "PVC", "PS"]

SMILESS = [
 "O=C(O)c1ccc(C(O)=O)cc1",
 "C=C",
 "ClC=C",
 "C=CC",
"c1ccccc1C=C",
"CC(Cl)CC(Cl)CC(Cl)CC(Cl)CC(Cl)CC(Cl)"
]


for SMILES in SMILESS:
    orca = OrcaWrapper(SMILES, theory='ri-mp2', basis='def2-tzvpp', parallel=4)

    with open(SMILES+".inp","w") as f:
        for line in orca:
            print(line, file= f)

    !cat $SMILES\.inp
    !LD_LIBRARY_PATH=/Users/yeren/orca /Users/yeren/orca/orca $SMILES\.inp
```
